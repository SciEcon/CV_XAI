```
### be sure to provide the raw link
import pandas as pd
import numpy as np
df_CM=pd.read_csv("https://raw.githubusercontent.com/coinmetrics-io/data/master/csv/btc.csv")
df_CM.head()
```

```
df_CM=df_CM[['time','BlkCnt','IssTotNtv','SplyCur','TxTfrValAdjUSD','CapMrktCurUSD','NVTAdj90','PriceUSD','FeeTotUSD','IssTotUSD','AdrActCnt']]
df_CM.head()
```

```
df_CM.info()
```

```
df_CM['time']=pd.to_datetime(df_CM['time'])
df_CM.info()
```

```
pd.Timestamp(df_CM['time'][0])
```

```
df_CM['time']=df_CM['time'].dt.strftime('%Y-%m-%d')
```

```
df_CM.set_index('time',drop=True,inplace=True)
```

```
df_CM.index.name=None
df_CM.head()
```
