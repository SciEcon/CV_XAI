```
### be sure to provide the raw link (click view raw to get the url)
import pandas as pd
import numpy as np
df_utxo=pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/Fintech_AI/master/data/utxo_age_distribution_2020_03_27.csv")
df_utxo.head()
```

```
df_utxo.info()
```

```
df_utxo=df_utxo.rename(columns={'<1d, x; 1d-1w, x; 1w-1m, x; 1-3m, x; 3-6m, x; 6-12m, x; 12-18m, x; 18-24m, x; 2-3y, x; 3-5y, x; >5y, x; BTC/USD, x':'date'})
```

```
df_utxo['date']=pd.to_datetime(df_utxo['date'])
df_utxo.info()
```

```
pd.Timestamp(df_utxo['date'][0])
```

```
df_utxo['date']=df_utxo['date'].dt.strftime('%Y-%m-%d')
```

```
df_utxo.set_index('date',drop=True,inplace=True)
```

```
df_utxo.index.name=None
df_utxo.head()
```

```
df_utxo.to_csv('UTXO_2020_07_03.csv')
```
