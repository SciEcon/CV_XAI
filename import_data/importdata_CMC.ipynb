```
import pandas as pd
import numpy as np
df_CMC=pd.read_csv('CMC_origin_28_04_2013__24_06_2020.csv' )
df_CMC.head()
```

```
df_CMC.info()
```

```
df_CMC['Date']=pd.to_datetime(df_CMC['Date'])
df_CMC.info()
```

```
pd.Timestamp(df_CMC['Date'][0])
```

```
df_CMC['Date']=df_CMC['Date'].dt.strftime('%Y-%m-%d')
```

```
df_CMC.head()
```

```
df_CMC.set_index('Date',drop=True,inplace=True)
```

```
df_CMC.index.name=None
df_CMC.head()
```

```
df_CMC['Volume']=df_CMC['Volume'].str.replace(',', '')
df_CMC.head()
```

```
df_CMC['Volume']=pd.to_numeric(df_CMC['Volume'])
```

```
df_CMC['Type']='BTC'
```

```
df_Volume = df_CMC[['Volume','Type']]
```

```
df_Volume.head()
```

```
###reverse the dataframe: https://www.marsja.se/six-ways-to-reverse-pandas-dataframe/
df_Volume = df_Volume.sort_index(ascending=True)
df_Volume.head(300)
```

```
import matplotlib.pyplot as plt
df_Volume['Volume'].plot()
```

```
df_Volume.to_csv('Volume.csv')
```

```
###Time Series with Range Selector Buttons
fig = px.line(df_Volume, x=df_Volume.index, y='Volume')
fig.update_layout(xaxis_range=['2013-04-28','2020-06-24'],
                  title_text="BTC Volume: from 2013-04-28 to 2020-06-24")
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```
###Customizing Tick Label Formatting by Zoom Level
fig = px.line(df_Volume, x=df_Volume.index, y='Volume')
fig.update_layout(xaxis_range=['2013-04-28','2020-06-24'],
                  title_text="BTC Volume: from 2013-04-28 to 2020-06-24")
fig.update_xaxes(
    rangeslider_visible=True,
    tickformatstops = [
        dict(dtickrange=[None, 1000], value="%H:%M:%S.%L ms"),
        dict(dtickrange=[1000, 60000], value="%H:%M:%S s"),
        dict(dtickrange=[60000, 3600000], value="%H:%M m"),
        dict(dtickrange=[3600000, 86400000], value="%H:%M h"),
        dict(dtickrange=[86400000, 604800000], value="%e. %b d"),
        dict(dtickrange=[604800000, "M1"], value="%e. %b w"),
        dict(dtickrange=["M1", "M12"], value="%b '%y M"),
        dict(dtickrange=["M12", None], value="%Y Y")
    ]
)
fig.show()
```

```
###Time Series with Range Selector Buttons
fig = px.line(df_Volume, x=df_Volume.index, y='Volume')
fig.update_layout(xaxis_range=['2013-04-28','2020-06-24'],
                  title_text="BTC Volume: from 2013-04-28 to 2020-06-24")
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```
